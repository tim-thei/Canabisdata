```python
import pandas as pd
import altair as alt
from vega_datasets import data
```

```python
#Timeseries regarding Canabis Legalisation in the US
#Self created dataset
#informations from: https://www.usnews.com/news/best-states/slideshows/where-is-pot-legal
#state ids are based on ANSI standardized code: https://www2.census.gov/geo/docs/reference/state.txt
df=pd.read_csv('LegalisationTS.csv', sep=';')

#varlist=['id', 'state', 'year', 'legal']

#Map of the US
states = alt.topo_feature(data.us_10m.url, feature='states')

df
```

```python
select_year = alt.selection_single(
    name='select', fields=['year'], init={'year': 2011},
    bind=alt.binding_range(min=2011, max=2022, step=1)
)
```

```python
alt.Chart(states, title='US-States where canabis consume is legal').configure_title(fontSize=28).mark_geoshape(stroke='white').encode(
    color=alt.Color('legal:N', scale=alt.Scale(
            domain=['Yes', 'No'],
            range=['#27B521', '#DC381F'])),
    tooltip=['state:N', 'legal:N']
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(df, 'id', list(df.columns))
).properties(
    width=1000,
    height=600
).project(
    type='albersUsa'
).add_selection(select_year).transform_filter(select_year)
```
